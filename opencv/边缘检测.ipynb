##  边缘检测
[Opencv之图像边缘检测：1.Sobel算子(cv2.Sobel)](https://blog.csdn.net/qq_49478668/article/details/123609610)

![](../img/sobel.jpg)

```python
import cv2
import numpy as np
from l_utils import plt_show_bgr
```

```python
img_hp = cv2.imread('../img/hb.jpg')
img_ln = cv2.imread('../img/lena.jpg')
img_ln_gray = cv2.imread('../img/lena.jpg', cv2.IMREAD_GRAYSCALE)
img_xq = cv2.imread('../img/xq.jpg')
img_xq_gray = cv2.imread('../img/xq.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
sobelx = cv2.Sobel(img_hp, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img_hp, cv2.CV_64F, 0, 1, ksize=3)
sobelx = cv2.convertScaleAbs(sobelx)
sobely = cv2.convertScaleAbs(sobely)
plt_show_bgr(
    (img_hp, sobelx, sobely),
    ('img_hp', 'sobelx', 'sobely')
)
```

分别计算x和y再求和效果更好

```python
sobelxy = cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)
sobelxy_ = cv2.Sobel(img_hp, cv2.CV_64F, 1, 1, ksize=3)
sobelxy_ = cv2.convertScaleAbs(sobelxy_)
plt_show_bgr(
    (img_hp, sobelxy, sobelxy_),
    ('img_hp', 'sobelxy', 'sobelxy_')
)
```

```python
sobelx = cv2.Sobel(img_ln_gray, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img_ln_gray, cv2.CV_64F, 0, 1, ksize=3)
sobelxy_ = cv2.Sobel(img_ln_gray, cv2.CV_64F, 1, 1, ksize=3)

sobelx = cv2.convertScaleAbs(sobelx)
sobely = cv2.convertScaleAbs(sobely)
sobelxy_ = cv2.convertScaleAbs(sobelxy_)

sobelxy = cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)

plt_show_bgr(
    (img_ln_gray, sobelxy, sobelxy_),
    ('img_ln_gray', 'sobelxy', 'sobelxy_'),
    'gray'
)
```

## scharr 与 lapkacian算子
[Opencv之图像边缘检测：2.Scharr算子（cv2.Scharr）](https://blog.csdn.net/qq_49478668/article/details/123742957)

[Opencv之图像边缘检测：3.Laplacian算子（cv2.Laplacian）](https://blog.csdn.net/qq_49478668/article/details/123742957)

```python
scharrx = cv2.Scharr(img_ln_gray, cv2.CV_64F, 1, 0)
scharry = cv2.Scharr(img_ln_gray, cv2.CV_64F, 0, 1)
scharrx = cv2.convertScaleAbs(scharrx)
scharry = cv2.convertScaleAbs(scharry)

laplacian = cv2.Laplacian(img_ln_gray, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)


plt_show_bgr(
    (img_ln_gray, sobelxy, scharrx, scharry, laplacian),
    ('img_ln_gray', 'sobelxy', 'scharrx', 'scharry', 'laplacian'),
    'gray'
)
```

## Canny 边缘检测
[Opencv之图像边缘检测：4.Canny边缘检测（cv2.Canny）](https://blog.csdn.net/qq_49478668/article/details/123965246)

```python
edges1 = cv2.Canny(img_xq_gray, 32, 170)
edges2 = cv2.Canny(img_ln_gray, 50, 100)
edges3 = cv2.Canny(img_ln_gray, 80, 150)
plt_show_bgr(
	(img_xq_gray, edges1, img_ln_gray, edges2, edges3),
	('img_xq_gray', 'edges1', 'img_ln_gray', 'edges2', 'edges3'),
	gray='gray')
```
